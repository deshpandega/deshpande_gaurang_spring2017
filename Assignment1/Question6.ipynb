# Question 6
#### Under Inaugural Address package
#### For every president, find out how many unique words they used in speech. 
#### We will just assume that President Bush means one person


- Import statements for nltk library, string implementations and iteration implementations

```python
import nltk
import string
import itertools
from nltk.corpus import stopwords, inaugural as ing

stop = stopwords.words('english')      #to remove stop words
punc = string.punctuation              #to remove punctuations
```

- Find out all the unique words that each president used in his own speeches 

```python
speech_pres = {}
for file in ing.fileids():
    #Check each file and find out all the words used in each file except stop words and punctuations
    words = list(set(nltk.FreqDist(word.lower() for word in ing.words(file) if word.lower() not in stop and word.lower() not in punc)))
    pres = file[5:-4]                      #Find the president name who used these words
    if pres in speech_pres:                
        speech_pres[pres] += set(words)    #If president name as key is already present, extend the list of words in dict
    else:
        speech_pres[pres] = words          #Else, add those words in dict with president name as key
```

- Find out all words one president used and not used by all other presidents

```python
uniq_words = {}
for (key, value) in speech_pres.items():
    #Maintain a list of words used by current president in context
    list_APres = value
    
    #Maintain another list of words used by all other presidents and make a single list of words
    list_OtherPres = list(itertools.chain.from_iterable([valueLower for (keyLower, valueLower) in speech_pres.items() if key != keyLower]))
    
    #Find out unique words for current president in context
    uniq = [word for word in list_APres if word not in list_OtherPres]
    
    #Add these words in a dictionary with president name as key and list of all unique words as value
    if key not in uniq_words:
        uniq_words[key] = uniq
    
    #Print list of presidents along with unique words they used in their inagural speech
    print('President : ',key,'\nThe unique words he used are: \n', uniq_words[key],'\n','************************'*5,'\n')
```
