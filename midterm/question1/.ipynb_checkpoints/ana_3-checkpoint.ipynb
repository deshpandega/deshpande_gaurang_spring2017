```python
#Import statements 
import json, os, glob, email, fnmatch, string, nltk, re, matplotlib.pyplot as plt, operator
from email.message import EmailMessage
```

```python
#Fetch current dir using relative path and fetch enron data
current_dir = os.path.dirname('__file__')
data_dir = os.path.join(current_dir, '..', 'data','enron')

matches = []
for root, dirnames, filenames in os.walk(data_dir):
    for filename in fnmatch.filter(filenames, '*'):
        matches.append(os.path.join(root, filename))
```

```python
#Getting all the emails from all the users
message_of_interest=[]
for filename in matches:
    with open(filename,'r') as f:
        data_from_file =  f.read()
        message = email.message_from_string(data_from_file)
        message_of_interest.append(message)
```

```python
#Get the users who the mails belong. If the name is not found then use the email id for them
dictionary_for_users = {}
for message in message_of_interest:
    if message['X-Origin'] != None:
        if message['X-Origin'].title() not in dictionary_for_users:
            dictionary_for_users[message['X-Origin'].title()] = 1
        else:
            dictionary_for_users[message['X-Origin'].title()] += 1
    else:
        if message['From'] not in dictionary_for_users:
            dictionary_for_users[message['From']] = 1
        else:
            dictionary_for_users[message['From']] += 1
```

```python
#Sort the users according to their email usage
sorted_users_dict = sorted(dictionary_for_users.items(), key=operator.itemgetter(1), reverse=True)
```

```python
#Function to creaate folder to save output using relative path
def create_directory_for_output():
    current_dir = os.path.dirname('__file__')
    data_folder = os.path.join(current_dir, 'ana_3')
    if not os.path.exists(data_folder):
        os.mkdir(data_folder)
    return data_folder

output_folder=create_directory_for_output()
```

```python
# x and y axes and users for lables
x_val = []
y_val = []
users = []
count = 0
for value in sorted_users_dict[:30]:
    users.append(value[0])
    count+=1
    x_val.append(count)
    y_val.append(value[1])

fig = plt.figure()                                                           #Width of each bar
plt.bar(x_val, y_val, color='#9E9E9E')                                       #Graph type with values for both axes
plt.grid(True)                                                               #Show grid
plt.xticks(x_val, users,  rotation=70)
plt.ylabel('Activity')
plt.xlabel('Users')
plt.title('Top 30 Most Active Users')  
fig.set_figwidth(12, 10)                                                     #Graph width
plt.show()                                                                   ##Show graph
```

```python
file_name = 'most_active_users'
file_path = os.path.join(output_folder, file_name)
file_path+='.png'

#Save the plot of top 30 most active users from enron
fig.savefig(file_path, bbox_inches='tight')   # save the figure to file
```
